```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.describe()
```

**Checking Missing Values**

```python
df.isnull().sum()
```

**Remove Null Values**

```python
AgeMedian=df['Age'].median()
df['Age'].fillna(AgeMedian, inplace=True)
df['Age'].isnull().sum()
```

```python
CabinCount=df['Cabin'].count()
df['Cabin'].fillna(CabinCount, inplace = True)
df['Cabin'].isnull().sum()
```

```python
EmbarkedCount=df['Embarked'].count()
df['Embarked'].fillna(EmbarkedCount, inplace = True)
df['Embarked'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['PassengerId','Pclass','SibSp','Cabin','Parch','Cabin'])
df
```

```python
df['Age'].max()
```

```python
df['Age']=df['Age'].apply(lambda x: x if x < 100 else AgeMedian)
df['Age']
```

```python
df
```

```python
FareMedian=df['Fare'].median()
df['Fare'].fillna(FareMedian, inplace=True)
df['Fare'].isnull().sum()
```

```python
df['Fare']=df['Fare'].apply(lambda x: x if x < 30.0000 else FareMedian)
```

```python
df['Fare']=df['Fare'].apply(lambda x: x if x > 8.05 else FareMedian)
```

```python
df['Fare']
```

```python
df['Age']=df['Age'].apply(lambda x: x if x > 18 else AgeMedian)
df['Age']
```

```python
df['Sex']=df['Sex'].str.title()
df['Sex']
```

```python
df['Sex'] = df['Sex'].replace({'Female': 'F', 'Male': 'M'})
df['Sex']
```

```python
df['Survived'] = df['Survived'].replace({0: 'Death', 1: 'Live'})
df['Survived']
```

```python
df['Embarked'] = df['Embarked'].replace({'S': 'Southampton', 'C': 'Cherbourg','Q':'Queenstown'})
df['Embarked']
```

```python
df
```

```python
FareGender=df.groupby('Sex')['Fare'].mean()
FareGender
```

```python
HighFare=df[df['Fare'] > 16.000000]
HighFare
```

```python
df.to_csv('Cleaned_Titanic.csv',index=False)
```

```python
df["Age"] = df["Age"].astype(int)
df['Fare']=df['Fare'].astype(float)
df
```
