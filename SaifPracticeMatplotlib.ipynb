```python
x=np.array([1,4,7,9])
y=np.array([4,5,8,9])
font1={'family':'arial','color':'blue','size':30}
font2={'family':'serif','color':'red','size':20}
font3={'family':'serif','color':'green','size':20}
plt.plot(x,y)
plt.title("AI&ML", fontdict=font1)
plt.xlabel("AI",fontdict=font2)
plt.ylabel("ML",fontdict=font3)
plt.show()
```
