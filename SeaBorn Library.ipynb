#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') #do not need to bring dataset library have data
tips.head() #display first few/5 records
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='red')
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Histogram of Total Bills")
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True, color='pink', edgecolor='purple')
plt.title("Histogram of Total Bills")
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True, color='yellow',stat='density',edgecolor='red')
sns.kdeplot(tips['total_bill'], color='darkgreen', linewidth=2)
plt.title("Histogram of Total Bills")
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.title("Box plot of Tips by Day")
plt.show()
```

```python
sns.pairplot(tips) # imp
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='total_bill',y='tip', data=tips, color='yellow') 
plt.title("Customized Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
tips['bill_size'] = tips['total_bill'].apply(lambda x: 'High' if x > 20 else 'Low')
sns.set_style("whitegrid")
sns.scatterplot(
    x='total_bill',
    y='tip',
    data=tips,
    hue='bill_size',         
    palette={'High': 'red', 'Low': 'green'}
)
plt.title("Customized Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Create Voilin plot
plt.Figure(figsize=(10,6))
sns.violinplot(x='day', y='tip',data=tips, hue='time', split=True)
plt.title("Voilon Plot of Total Bills by Day and Tip")
plt.xlabel("Day of the week")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create Bar plot
plt.Figure(figsize=(10,6))
# sns.barplot(x='day', y='tip',data=tips, hue='sex',palette='pastel', ci=None)
sns.barplot(x='day', y='tip',data=tips, hue='sex',palette='pastel', errorbar=None)
plt.title("Average Tip Amount by Day and Sex")
plt.xlabel("Day of the week")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create a countplot
plt.Figure(figsize=(10,6))
sns.countplot(x='day', hue='smoker', data=tips, palette='coolwarm')
plt.title("Count of Dinners by Day and Smoking Preferences")
plt.xlabel("Day of the Week")
plt.ylabel("Count of Dinners")
plt.legend(title='Smoker')
plt.show()
```
