```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Cleaned_titanic.csv')
df
```

```python
df.isnull().sum()
```

```python
x= df['Age']
y= df['Fare']
```

```python
plt.title('Titanic')

plt.xlabel('Age' )
plt.ylabel('Fare')

plt.scatter(x,y, color='red' , label='Age')
plt.scatter(x,y, color='blue' , label='Fare' , alpha=0.5)
plt.legend()
plt.show()
```

```python
x= df['Age']
y= df['Fare']
plt.title('Titanic')
plt.xlabel('Age' )
plt.ylabel('Fare')

plt.scatter(x,y, edgecolor='red' , facecolor='green')
#plt.scatter(x,y, color='blue' , label='Fare' , alpha=0.5)
#plt.legend()
plt.show()
```

```python
x= df['Age']
y= df['Fare']

plt.title('Titanic')
plt.xlabel('Age' )
plt.ylabel('Fare')

colors= ["red" if age >30 else 'green' for age in x]
plt.scatter(x,y, c=colors)
plt.show()
```

```python
x= df['Age']
y= df['Fare']

plt.title('Titanic')
plt.xlabel('Age' )
plt.ylabel('Fare')
colors= ["red" if age <=10 else
         'green' if age <=20 else
         'blue' if age <=30 else
         'black' if age <=40 else
         'pink' if age <=50 else
         'orange' if age <=60 else
         'violet' if age <=70 else
          'gray'for age in x]

plt.scatter(x,y, c=colors , marker='+')
plt.show()
```
