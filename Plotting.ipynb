```python
import pandas as pd

df = pd.read_csv("Cleaned_data.csv") 
median_salary = df["Salary"].median()

df.loc[df["Salary"] > 75000, "Salary"] = median_salary

df.to_csv("Cleaned_Salary_Data.csv", index=False)
print("Cleaned file saved as 'Cleaned_Salary_Data.csv'")
```

```python
x=df['Age']
y=df['Salary']
plt.plot(x,y, marker='o') 

for i in range(len(df)):
    if df['Salary'][i] > 50000:
        plt.plot(x[i], y[i], marker='o', color='red')
    else:
        plt.plot(x[i], y[i], marker='o', color='blue')

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('AI & ML')
plt.show()
```
