```python
import numpy as np
import matplotlib.pyplot as plt  # Correct import

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, label="Line 1")
plt.plot(y2, label="Line 2")

plt.title("AI & ML")
plt.xlabel("AI")  # Correct function call
plt.ylabel("ML")  # Correct function call

plt.show()
```

```python
x=np.array([20,25,30,35,40])
y=np.array([5,10,15,20,25])
font1={'family':'arial','color':'blue','size':20}
font2={'family':'serif','color':'red','size':15}
plt.title("Sports Watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorise burnage",fontdict=font2)

plt.plot(x,y)
plt.show()
```
