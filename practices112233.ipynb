```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cleanbybilal_titanic.csv')
df
```

```python
x=df['Age']
y=df['Fare']
plt.scatter(x,y)
```

```python
x=df['Age']
y=df['Fare']
plt.xlabel('Age',color='red')
plt.ylabel('fare',color='green')
plt.scatter(x,y)
plt.scatter(x,y,c='red',label='Age')
plt.scatter(x,y,c='green',label='fare',alpha=0.5)
plt.legend(['Age(red)','Fare(green)'])

plt.show()
```

```python
x=df['Age']
y=df['Fare']
plt.xlabel('Age',color='red')
plt.ylabel('fare',color='green')
colors=['red' if age>30 else 'green' for age in x]
plt.scatter(x,y,c='red',label='Age')
plt.scatter(x,y,c='green',label='fare',alpha=0.5)
plt.legend(['Age(red)','Fare(green)'])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
x=df['Age']
y=df['Fare']
plt.title('Titanic')
plt.xlabel('Age')
plt.ylabel('Fare')
colors=['red' if age<=10 else 
        'green' if age<=20 else
        'yellow' if age<=30 else
        'pink' if age<=40 else
        'gray' if age<=50 else 
        'blue' if age<=60 else 
        'orange' if age<=70 else 
        'black' for age in x]
plt.scatter(x,y,c=colors, marker="*")
plt.show()
```
