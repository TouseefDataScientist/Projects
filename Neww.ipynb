```python
df=pd.read_csv('cleaned_titanic.csv')
```

```python
df
```

```python
x=df['Age']
```

```python
y=df['Fare']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, c='purple', label='Age')
plt.scatter(x,y, c='green', label='Fare', alpha=0.5)

plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.legend(['Age(red)', 'Fare(green)'])
plt.show()
```
