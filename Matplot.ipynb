```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([3,4,5,6,7,8,9]) 
plt.scatter(x,y ,color='hotpink')

x=np.array([9,10,11,12,13,14,15])
y=np.array([5,12,3,9,7,4,5]) 
plt.scatter(x,y ,color='blue')

plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([3,4,5,6,7,8,9])
colors = np.array(["red","green","blue","hotpink","orange","black","yellow"])
plt.scatter(x,y , c=colors)
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([3,4,5,6,7,8,9])
colors = np.array(["red","green","blue","hotpink","orange","black","yellow"])
plt.scatter(x,y , c=colors)
plt.colorbar()
plt.show()
```
