```python
df=pd.read_csv('Titanic clean data.csv')
df
```

```python
x=df['Age']
```

```python
y=df['Fare']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.title("Titanic Clean Data")
plt.xlabel("Age")  
plt.ylabel("Fare")
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, c='green', label='Age')
plt.scatter(x,y, c='yellow', label='Fare', alpha=0.5)

plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.legend(['Age(green)', 'Fare(yellow)'])
plt.show()
```

```python
plt.scatter(x,y, edgecolors='orange', facecolor='purple')
plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
x=df['Age']
y=df['Fare']
colors=['red'if Age >30 else 'green' for Age in x]
plt.scatter(x,y,c=colors)
plt.show()
```

```python
x=df['Age']
y=df['Fare']
colors=[
    'red'if 0<age<=10
    else 'green' if 11<age<=20
    else 'yellow'if 20<age<=30
    else 'brown'if 30<age<=40
    else 'orange'if 40<age<=50
    else 'purple'if 50<age<=60
    else 'black'if 60<age<=70 
    else 'blue'
    for age in x
]
plt.scatter(x,y,c=colors,marker='+')
plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
import matplotlib.patches as mpatches

x = df['Age']
y = df['Fare']

colors = [
    'red' if 0 < age <= 10 
    else 'green' if 11 < age <= 20 
    else 'yellow' if 20 < age <= 30 
    else 'brown' if 30 < age <= 40 
    else 'orange' if 40 < age <= 50 
    else 'purple' if 50 < age <= 60 
    else 'black' if 60 < age <= 70  
    else 'blue'  
    for age in x
]

plt.scatter(x, y, c=colors, marker='+')
plt.title("Titanic Data")
plt.xlabel("Age")
plt.ylabel("Fare")

# Creating a custom legend
legend_patches = [
    mpatches.Patch(color='red', label='0-10'),
    mpatches.Patch(color='green', label='11-20'),
    mpatches.Patch(color='yellow', label='20-30'),
    mpatches.Patch(color='brown', label='30-40'),
    mpatches.Patch(color='orange', label='40-50'),
    mpatches.Patch(color='purple', label='50-60'),
    mpatches.Patch(color='black', label='60-70'),
    mpatches.Patch(color='blue', label='Others'),
]

plt.legend(handles=legend_patches, title="Age Groups")

plt.show()
```
