**Marker Description**
' o ' Circle 
' * ' Star 
' . ' Point 
' , ' Pixel 
' x ' X
' X ' X ( filled ) <br>
' + ' Plus 
' P ' Plus ( filled ) 
' s ' Square
' D ' Diamond 
' d ' Diamond ( thin ) <br>
' p ' Pentagon
' H ' Hexagon 
' h ' Hexagon 
' v ' Triangle Down 
' ^ ' Triangle Up <br>
' < ' Triangle Left 
' > ' Triangle Right 
' 1 ' Tri Down 
' 2 ' Tri Up 
' 3 ' Tri Left <br>
' 4 ' Tri Right
' | ' Vline 
' _ ' Hline

```python
y1=np.array([3,8,1,10])
y2=np.array([1,2,7,11])
plt.plot(y1,y2)
plt.xlabel("ai")
plt.ylabel("ml")
plt.show()
```

```python
y1=np.array([3,8,1,10])
y2=np.array([1,2,7,11])
plt.plot(y1,y2)
plt.title("AI & ML")
plt.xlabel("ai")
plt.ylabel("ml")
plt.show()
```

```python
x=np.array([20,25,30,35,40])
y=np.array([5,10,15,20,25])
font1={'family':'arial','color': 'blue','size':20}
font2={'family':'serif','color':'red','size':15}
plt.title("sports watch data",fontdict=font1)
plt.xlabel("Average Puls",fontdict=font2)
plt.ylabel("calories burned",fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([10,20,30,40,50])
y=np.array([50,60,70,80,90])
plt.plot(x, y)
plt.title("AI & ML Class")
plt.xlabel("Salary")
plt.ylabel("Age")
plt.grid()
plt.show()
```

```python
x=np.array([10,20,30,40,50])
y=np.array([50,60,70,80,90])
plt.plot(x, y,linewidth=6)
plt.title("AI & ML Class")
plt.xlabel("Salary")
plt.ylabel("Age")
plt.grid()
plt.grid(color='green',linestyle='--',linewidth=1)
plt.show()
```
